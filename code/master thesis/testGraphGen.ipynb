Genero una matrice $n^2\times n^2$ e vi inserisco il valore di $v$ dove ho un arco per la corrispondente riga

```python
def set_speed(graph, n, v):
    V = [[0 for j in range(n*n)]for i in range(n*n)]
    for ind in graph.keys():
        for neigh in graph[ind]:
            V[ind][neigh] = v
    return V
```

```python
def set_distance(graph, n, d):
    D = [[0 for j in range(n*n)]for i in range(n*n)]
    for ind in graph.keys():
        for neigh in graph[ind]:
            D[ind][neigh] = d
    return D
```

Con costo unitario, utilizza BFS

```python
def shortest_path(graph, start ,destination):
	explored = []
	
	queue = [[start]]
	
	if start == destination:
		print("Same Node")
		return
	while queue:
		path = queue.pop(0)
		node = path[-1]
		if node not in explored:
			neighbours = graph[node]
			for neighbour in neighbours:
				new_path = list(path)
				new_path.append(neighbour)
				queue.append(new_path)
				
				if neighbour == destination:
					return new_path
			explored.append(node)
	return
```

genera volo:
scelgo due nodi a caso ma differenti, tra di loro calcolo cammino minimo

$k$ numero di voli, $n$ numero nodi per riga, grafo e seme

```python
def generate_flights(k, n, graph, seed):
    F = []
    rand.seed(seed)
    for i in range(k):
        s = rand.randint(0,n*n-1)
        e = s
        while e==s:
            e = rand.randint(0,n*n-1)
        F.append(shortest_path(graph, s, e))
    #print(F)
    return F
```

gen conflict
per ogni coppia di voli, cerco voli che hanno delle intersezioni, e le salvo in conf, identificato da valore volo, nodi interessati e distanza minima da rispettare

```python
def get_conflicts(F, D, v,n): #very specific to grid and metroplex
    conf = []
    tau = D/v 
    for i in range(len(F)):
        for j in range(i):
            Fi, Fj = set(F[i]), set(F[j])
            #intersection of two flights
            tmp = Fi & Fj
            for c in tmp:
                conf.append([i,j,c, tau])
    return conf


def matrix_to_string(data):
    tmp = ""
    for i in range(len(data)):
        for x in data[i]:
            tmp += str(x) + " "
        tmp += '\n'
    return tmp

def matrix_to_string_ID(data):
    tmp = ""
    for i in range(len(data)):
        tmp += str(i) + " "
        for x in data[i]:
            tmp += str(x) + " "
        tmp += '\n'
    return tmp
```

## Metroplex


Genero grafo griglia, ne creo una copia spostata di $n\times n$, infine connetto con 2 archi

primo arco connette nodo prima colonna ultima riga a prima riga prima colonna della nuova griglia
secondo arco connette nodo ultima colonna ultima riga a prima riga prima colonna della nuova griglia

```python
def generate_network(n):
    edges = []
    for i in range(n):
        for j in range(n):
            if j != n-1:
                edges.append([j+i*n, j+i*n+1])
            if i != n-1:
                edges.append([j+i*n, j+(i+1)*n])

    #creo seconda griglia prendendo la prima e aggiungendo n*n
    tmp = []
    for [i,j] in edges: #second grids
        tmp.append([i+n*n, j+n*n])
    edges += tmp


    # primo ultima riga e prima colonna 
    edges += [[n*(n-1), n*n]] #connection between grids
    #primo ultima riga e ultima colonna
    edges += [[n*(n-1)+n-1, n*n+n-1]]
    graph = defaultdict(list)
    for edge in edges:
        a, b = edge[0], edge[1]
         
        graph[a].append(b)
        graph[b].append(a)
    return graph

generate_network(3)
```

## Airport

Creo n nodi e li collego a nodo hub $n$

```python

def generate_network(n):
    edges = []
    for i in range(n):
        edges.append([i,n])
    
    graph = defaultdict(list)
    for edge in edges:
        a, b = edge[0], edge[1]
         
        graph[a].append(b)
        graph[b].append(a)
    return graph
```

Come i precedenti

```python


def set_speed(graph, n, v):
    V = [[0 for j in range(n+1)]for i in range(n+1)]
    for ind in graph.keys():
        for neigh in graph[ind]:
            V[ind][neigh] = v
    return V

def set_distance(graph, n, d):
    D = [[0 for j in range(n+1)]for i in range(n+1)]
    for ind in graph.keys():
        for neigh in graph[ind]:
            D[ind][neigh] = d
    return D

def shortest_path(graph, start ,goal):
	explored = []
	
	queue = [[start]]
	
	if start == goal:
		print("Same Node")
		return
	while queue:
		path = queue.pop(0)
		node = path[-1]
		
		if node not in explored:
			neighbours = graph[node]
			
			for neighbour in neighbours:
				new_path = list(path)
				new_path.append(neighbour)
				queue.append(new_path)
				
				if neighbour == goal:
					return new_path
			explored.append(node)
	return

def generate_flights(k, n, graph, seed):
    F = []
    rand.seed(seed)
    for i in range(k):
        s = rand.randint(0,n-1)
        e = s
        while e==s:
            e = rand.randint(0,n-1)
        F.append(shortest_path(graph, s, e))
    #print('flights', F)
    return F
```

```python
def get_conflicts(F, D, v, n): # specific to airport
    conf = []
    tau = D/v 
    for i in range(len(F)):
        for j in range(i):
            Fi, Fj = set(F[i]), set(F[j])
            tmp = Fi & Fj
            for c in tmp:
                if (i-j)/float(n) < float(1)/float(2):
                    #print((i-j)/float(n)*math.pi)
                    S = 1/math.sin((i-j)/float(n)*math.pi)
                else:
                    S = 1
                tau = S*D/v
                conf.append([i,j,c, tau])
    return conf
```
