```python
listFree = [0,28]
modified = []
with open('APmetroplex4_Fixed_fix.run', 'r') as file:
    righe = file.readlines()  # Ogni riga  un elemento della lista `righe`

for rigaNormale in righe:
    riga = rigaNormale.split(" ")
    if (riga[1].split("[")[0] == "t_down_fixed") and (int(riga[1].split("[")[1].split(",")[0]) in listFree):
        print("fix t_down[" +riga[1].split("[")[1]+riga[2])
        modified.append(str("fix t_down[" +riga[1].split("[")[1]+riga[2]))

    elif (riga[1].split("[")[0] == "t_up_fixed") and (int(riga[1].split("[")[1].split(",")[0]) in listFree):
        print("fix t_up[" +riga[1].split("[")[1]+riga[2])
        modified.append(str("fix t_up[" +riga[1].split("[")[1]+riga[2]))

    elif (riga[1].split("[")[0] == "t_ear_fixed") and (int(riga[1].split("[")[1].split(",")[0]) in listFree):
        print("fix t_ear[" +riga[1].split("[")[1]+riga[2])
        modified.append(str("fix t_ear[" +riga[1].split("[")[1]+riga[2]))

    elif (riga[1].split("[")[0] == "t_lat_fixed") and (int(riga[1].split("[")[1].split(",")[0]) in listFree):
        modified.append(str("fix t_lat[" +riga[1].split("[")[1]+riga[2]))
        print(str("fix t_lat[" +riga[1].split("[")[1]+riga[2]))

    elif riga[1].split("[")[0] == "pass2Fixeds":
        continue
    elif riga[1].split("[")[0] == "wFixed":
            if int(riga[1].split("[")[1].split(",")[2].split("]")[0]) in listFree:
                modified.append(str("fix w[" +riga[1].split("[")[1]+riga[2]))
                print(str("fix w[" +riga[1].split("[")[1]+riga[2]))
            else : 
                continue
    else:
        print(riga[0]+" "+riga[1]+" "+riga[2])
```
